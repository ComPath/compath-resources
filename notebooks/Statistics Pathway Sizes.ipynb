```python
from bio2bel_kegg import Manager as KeggManager
from bio2bel_reactome import Manager as ReactomeManager
from bio2bel_wikipathways import Manager as WikiPathwaysManager

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy
```

```python
%matplotlib inline
```

```python
kegg_manager = KeggManager()
reactome_manager = ReactomeManager()
wikipathways_manager = WikiPathwaysManager()
```

#### Get Pathway Distributions

```python
kegg_distribution = kegg_manager.get_pathway_size_distribution()
reactome_distribution = reactome_manager.get_pathway_size_distribution()
wikipathways_distribution = wikipathways_manager.get_pathway_size_distribution()
```

#### Plot Distributions

```python
fig = plt.figure(figsize=(25, 6))

keggs = list(kegg_distribution.values())
reactomes = list(reactome_distribution.values())
wikipathways = list(wikipathways_distribution.values())
             
ax = fig.add_subplot(1, 3, 1)
ax.set_title('Distribution KEGG pathways size', fontsize=18)
ax.set_xlabel('Size', fontsize=18)
ax.set_ylabel('Frequency', fontsize=18)
ax.set_xlim(0, 1300)
sns.distplot(keggs, bins=20, hist=False)

ax = fig.add_subplot(1, 3, 2)
ax.set_title('Distribution Reactome pathways size', fontsize=18)
ax.set_xlabel('Size', fontsize=18)
ax.set_ylabel('Frequency', fontsize=18)
ax.set_xlim(0, 2800)
sns.distplot(reactomes, bins=20, hist=False)

ax = fig.add_subplot(1, 3, 3)
ax.set_title('Distribution WikiPathways pathways size', fontsize=18)
ax.set_xlabel('Size', fontsize=18)
ax.set_ylabel('Frequency', fontsize=18)
ax.set_xlim(0, 400)
sns.distplot(wikipathways, bins=20, hist=False)

plt.show()
```

Figure 1. Pathway Size Distributions of KEGG, Reactome, WikiPathways. 

#### Statistics Summary

```python
print('Mean KEGG: {}'.format(scipy.stats.describe(keggs)))
print('Mean Reactome: {}'.format(scipy.stats.describe(reactomes)))
print('Mean WikiPathways: {}'.format(scipy.stats.describe(wikipathways)))
```
